```python
import torch
from transformers import AutoTokenizer, AutoModel
# Check if CUDA is available and use GPU if it is
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load the tokenizer and model, then move the model to the GPU
tokenizer = AutoTokenizer.from_pretrained("dbmdz/bert-base-italian-cased")
model = AutoModel.from_pretrained("dbmdz/bert-base-italian-cased").to(device)
model.eval()  # Set the model to evaluation mode
```

```python
def embed_texts(texts, tokenizer, model, device):
    inputs = tokenizer(texts, return_tensors='pt', padding=True, truncation=True).to(device)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1)  # Average the embeddings over all tokens
```

```python
from torch.nn.functional import cosine_similarity

def rank_documents(query, positives, negatives, tokenizer, model, device):
    # Embed the query
    query_embedding = embed_texts([query], tokenizer, model, device)

    # Embed all documents (positives and negatives)
    all_documents = positives + negatives
    doc_embeddings = embed_texts(all_documents, tokenizer, model, device)

    # Calculate cosine similarities
    similarities = cosine_similarity(query_embedding, doc_embeddings)

    # Rank documents by similarity (highest to lowest)
    ranked_indices = torch.argsort(similarities, descending=True)
    ranked_documents = [all_documents[i] for i in ranked_indices]

    return ranked_documents
```

```python
def calculate_r_at_1k(dataset, tokenizer, model, device):
    total_recall = 0
    for data in dataset:
        query = data['query']
        positives = data['positive']
        negatives = data['negative']

        ranked_documents = rank_documents(query, positives, negatives, tokenizer, model, device)

        # Top 1000 documents
        top_1000 = ranked_documents[:1000]

        # Number of relevant documents in the top 1000
        retrieved_relevant = sum(1 for doc in top_1000 if doc in positives)

        # Calculate recall for this query
        recall = retrieved_relevant / len(positives)
        total_recall += recall

    # Average recall across all queries
    r_at_1k = total_recall / len(dataset)
    return r_at_1k

def calculate_mrr_at_10(dataset, tokenizer, model, device):
    total_reciprocal_rank = 0
    for data in dataset:
        query = data['query']
        positives = data['positive']
        negatives = data['negative']

        ranked_documents_mrr = rank_documents(query, positives, negatives, tokenizer, model, device)

        # Top 10 documents
        top_10 = ranked_documents_mrr[:10]

        # Find the rank of the first relevant document
        for rank, doc in enumerate(top_10, start=1):
            if doc in positives:
                total_reciprocal_rank += 1 / rank
                break  # Only the first relevant document matters

    # Average reciprocal rank across all queries
    mrr_at_10 = total_reciprocal_rank / len(dataset)
    return mrr_at_10

# Example usage with your dataset
r_at_1k = calculate_r_at_1k(dataset['train'], tokenizer, model, device)
mrr_at_10 = calculate_mrr_at_10(dataset['train'], tokenizer, model, device)

print(f"R@1k: {r_at_1k}")
print(f"MRR@10: {mrr_at_10}")
```
